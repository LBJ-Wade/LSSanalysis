## Import/sanity check the land area data.

* Read in the map of land area and derive the longitude/latitude coordinates from of the maps from the header.
* Figured out the map formats from [here](http://sedac.uservoice.com/knowledgebase/articles/123459-i-downloaded-the-gpwv3-population-density-ascii-f)
* Compute the total area of each pixel from longitude/latitude as a sanity check, and compare against the land area map.
* Sanity check passed; found some funky pixels with more area than makes sense, but it's a tiny fraction of the total map so we can safely ignore it.
* Visualize: area per pixel decreases towards the poles as expected.  I see continents!

```python
## define a function to read in the header information from the maps and generate a grip of longitude and latitude points corresponding to the maps.
landareafname = "gpwv3_area_ascii_25/glareag.asc"


def getllfromheader(filename):
  import re
    
  ifp = open(filename,'r')
  ## 6 header lines.  
  hdat = {}
  for tag, fmt in zip(['ncols','nrows','xllcorner','yllcorner','cellsize','NODATA_value'],['int','int','float','float','float','int']):
    ll = ifp.readline()
    if not re.search(tag,ll):
      print 'header ',tag,'did not show up!'
      return None
    if fmt == 'int':
      hdat[tag] = int(ll.split(tag)[-1])
    elif fmt == 'float':
      hdat[tag] = float(ll.split(tag)[-1])
    else:
      hdat[tag] = ll.split(tag)[-1]
  ifp.close()
  ## - 180 to 180
  longitude1d = np.arange(0,hdat['ncols'],1)*hdat['cellsize'] + hdat['xllcorner'] + hdat['cellsize']*0.5 ## return central coordinate.
  latitude1d = np.arange(hdat['nrows']-1,-0.5,-1)*hdat['cellsize'] + hdat['yllcorner'] + hdat['cellsize']*0.5 ## return central coordinate.
  longitude, latitude = np.meshgrid(longitude1d, latitude1d)
  return longitude, latitude, hdat

longitude, latitude, hdat = getllfromheader(landareafname)
print 'latitude extent:',latitude[0,0], latitude[-1,0]
print 'longitude extent:',longitude[0,0], longitude[0,-1]

# read in the land area map.
# tricky!  the y coordinate will be longitude (vertical lines, usual x coordinate), x coordinate latitude (horizontal lines, usual y coordinate).
# I guess this makes sense as a format as an image, unlike 2d data arrays I'm used to working with.
agrid = np.loadtxt(landareafname,skiprows=6)

# sanity check against my calculation of total area of each pixel.
Rearthgoogle = 6378.1 # kilometers from google.com
mya = hdat['cellsize']*np.pi/180.*np.fabs(np.sin(np.radians(latitude-0.5*hdat['cellsize'])) - np.sin(np.radians(latitude+0.5*hdat['cellsize'])))*Rearthgoogle**2 ## result in steradians, convert to square kilometers using radius of earth.

# first look at the equator
xx = np.where(np.fabs(latitude) < 0.03)
print 'equator latitude:',latitude[xx].mean(), '+/-',latitude[xx].std()
print 'do we agree on km^2/pixel on equator?',mya[xx].mean(), agrid[xx].max()
print 'not quite.  Could be difference in the earth radius we assumed or non-spherical geometry of the earth (if they accounted for that?)'

# scan through rows of latitude.  Maximum of land area should be equal to the area I compute for each pixel.
# keep track of the ratio of the areas for each latitude.
rlist = []
for i in range(len(agrid[:,0])):
  #print latitude[i,0], latitude[i,-1], agrid[i,:].max(), mya[i,0], agrid[i,:].max()/mya[i,0]
  if np.fabs(agrid[i,:].max()/mya[i,0] - 1.) < 0.1: rlist.append(agrid[i,:].max()/mya[i,0])

rlist = np.array(rlist)
rlist.sort()
print 'median, min, max area ratios',rlist[len(rlist)/2],rlist[0],rlist[-1], rlist.mean()
print 'weird, why are there regions where their area is larger than the normal calculation?  how much area are we talking?'

## update myarea and Rearth.
Rearth = Rearthgoogle*(rlist[len(rlist)/2])**0.5
mya = hdat['cellsize']*np.pi/180.*np.fabs(np.sin(np.radians(latitude-0.5*hdat['cellsize'])) - np.sin(np.radians(latitude+0.5*hdat['cellsize'])))*Rearth**2

#update rlist
rlist = []
for i in range(len(agrid[:,0])):
  #print latitude[i,0], latitude[i,-1], agrid[i,:].max(), mya[i,0], agrid[i,:].max()/mya[i,0]
  if np.fabs(agrid[i,:].max()/mya[i,0] - 1.) < 0.1: rlist.append(agrid[i,:].max()/mya[i,0])

rlist = np.array(rlist)
rlist.sort()
print 'updated median, min, max area ratios',rlist[len(rlist)/2],rlist[0],rlist[-1], rlist.mean()

xx = np.where(agrid > mya*1.001)
print 'this much fractional area excess',mya[xx].sum()/mya.sum(), 'so can be safely ignored.'

plt.figure(figsize=[15,15*(latitude.min()-latitude.max())/(longitude.min()-longitude.max())])
ii=plt.imshow(agrid/agrid.max(),extent=[longitude.min(),longitude.max(),latitude.min(),latitude.max()])
#plt.colorbar(ii,orientation='horizontal')
print 'this is just a plot of the land area per pixel to see the distortion of the map projection.'
```

## One point distributions/histograms of population density
The cumulative histogram of population density reproduces several statistics in the internet.org video/whitepaper.

* 80% of humans live in "urban" environments, i.e., at least ~100 people per square kilometer.  Urban environments occupy only 9.6% of all land area.
* To provide coverage for 99% of the world population, we need to reach densities down to 4.2 people/sqr km and cover 46% of all land area.
* This problem ("internet for all") definitely obeys the 80/20 rule!  The next 19% will be more work and more costly; for a naive estimate that cost $\propto$ area covered, the cost per person for access would increase by a factor of 16 (or worse, terrain more difficult, less infrastructure like roads in place to lay fiber optic cable!)

```python
## now make some basic plots to understand how population is distributed statistically
## first get indices that aren't NODATA_value, then sort based on population density.
xgood = np.where(pdens != hdat['NODATA_value'])
pdensnew = (pdens[xgood]).flatten()
anew = (agrid[xgood]).flatten()
atotnew = (mya[xgood]).flatten()
pcntsnew = (pcnts[xgood]).flatten()
x = pdensnew.argsort()
pdensnew = pdensnew[x]
anew = anew[x]
pcntsnew = pcntsnew[x]
print 'total adjusted population in billions in year 2000:',pcntsnew.sum()/1.e9
print 'total area covered by this map chunk of lat/long:',mya.sum(), 'relative to expected',mya.sum()/(2.*np.pi*(np.sin(np.radians(latitude.max()))-np.sin(np.radians(latitude.min())))*Rearth**2)
print 'what fraction of the surface of the earth is land? %.1f percent' % (100.*anew.sum()/(4.*np.pi*Rearth**2))
xtmp = np.where(pdensnew > 0.)[0]
ytmp = np.where(pdensnew == 0.)[0]

print 'this much land area has population density 0:',anew[ytmp].sum(), 'square kilometers; fraction of total',anew[ytmp].sum()/anew.sum()
plt.figure(figsize=[5,5])
plt.loglog(pdensnew[xtmp],anew[xtmp].cumsum())
plt.xlabel('pop density',fontsize=16)
plt.ylabel('Area(pop density < p)',fontsize=16)
plt.show()
print 'wacky cumulative histogram, jumps probably correspond to discrete number of people/pixel or something.'
print 'probably does not matter, as most of the population lives at higher density, and the super low population density areas are a small fraction of the land area'
```

```python
## now let's try to reproduce internet.org video's claim that the world's population spans 5 orders of magnitude and 80% live in urban areas.
pfrac = [0.999,0.997,0.99,0.95,0.9,0.8,0.5]
pminlist = []
for ff in pfrac:
  xx = np.where(pcntsnew.cumsum()/pcntsnew.sum() > 1-ff)[0]
  pminlist.append(pdensnew[xx[0]])
  print 'min pop density to include %.1f percent of population: %.3f per square km' % (ff*100.,pdensnew[xx[0]])
  print 'this pop occupies this fraction of land (earth) area: %.3f (%.3f)' % (anew[xx[0]:].sum()/anew.sum(),anew[xx[0]:].sum()/(4.*np.pi*Rearth**2))
  print 'for this range, population density spans this many orders of magnitude: %.1f' % (np.log10(pdens.max()/pdensnew[xx[0]]))
  print ''

xx = np.where(pcntsnew.cumsum()/pcntsnew.sum() > 1-0.99)[0]
yy = np.where(pcntsnew.cumsum()/pcntsnew.sum() > 1-0.8)[0]
costfac = (anew[xx[0]:yy[0]].sum()/anew[yy[0]:].sum()/(pcntsnew[xx[0]:yy[0]].sum()/pcntsnew[yy[0]:].sum()))
print 'To go from 80% coverage to 99% coverage, cost per person using current technology (if internet access cost proportional to area covered) would increase by a factor of ','%.1f' % (costfac)
```

## Import National Identifier Grid
* Found code online to read the .dbf file containing the Country --> ID mapping.
* Wrote functions to select countries from the map and plot their population density/compute simple things.

```python
##oops, I have nothing to read dbf.  Copied some code off the web to open the dbf file:
## http://code.activestate.com/recipes/362715-dbf-reader-and-writer/

import struct, datetime, decimal, itertools

def dbfreader(f):
    """Returns an iterator over records in a Xbase DBF file.

    The first row returned contains the field names.
    The second row contains field specs: (type, size, decimal places).
    Subsequent rows contain the data records.
    If a record is marked as deleted, it is skipped.

    File should be opened for binary reads.

    """
    # See DBF format spec at:
    #     http://www.pgts.com.au/download/public/xbase.htm#DBF_STRUCT

    numrec, lenheader = struct.unpack('<xxxxLH22x', f.read(32))    
    numfields = (lenheader - 33) // 32

    fields = []
    for fieldno in xrange(numfields):
        name, typ, size, deci = struct.unpack('<11sc4xBB14x', f.read(32))
        name = name.replace('\0', '')       # eliminate NULs from string   
        fields.append((name, typ, size, deci))
    yield [field[0] for field in fields]
    yield [tuple(field[1:]) for field in fields]

    terminator = f.read(1)
    assert terminator == '\r'

    fields.insert(0, ('DeletionFlag', 'C', 1, 0))
    fmt = ''.join(['%ds' % fieldinfo[2] for fieldinfo in fields])
    fmtsiz = struct.calcsize(fmt)
    for i in xrange(numrec):
        record = struct.unpack(fmt, f.read(fmtsiz))
        if record[0] != ' ':
            continue                        # deleted record
        result = []
        for (name, typ, size, deci), value in itertools.izip(fields, record):
            if name == 'DeletionFlag':
                continue
            if typ == "N":
                value = value.replace('\0', '').lstrip()
                if value == '':
                    value = 0
                elif deci:
                    value = decimal.Decimal(value)
                else:
                    value = int(value)
            elif typ == 'D':
                y, m, d = int(value[:4]), int(value[4:6]), int(value[6:8])
                value = datetime.date(y, m, d)
            elif typ == 'L':
                value = (value in 'YyTt' and 'T') or (value in 'NnFf' and 'F') or '?'
            elif typ == 'F':
                value = float(value)
            result.append(value)
        yield result
        
bndyfname = "gpwv3_ntlbndid_ascii_25/bndsg.dbf"
f = open(bndyfname,'rb')
#bnames,btypes,bndy = list(dbfreader(f))
tmp = list(dbfreader(f))
bnames = tmp[0]
btypes = tmp[1]
bndy = tmp[2:]  #contents is a list [MapID, unknown, 3 letter abbreviation, UNID, country name']

## split off \n in the country names.
for i in range(len(bndy)):
  bndy[i][4] = bndy[i][4].strip()
    
## make two dictionaries going between UNID and MAPid
maptoUN = {}
UNtomap = {}
for i in range(len(bndy)):
  UNstr = '%03d' % (bndy[i][3])
  maptoUN[bndy[i][0]] = UNstr
  UNtomap[UNstr] = bndy[i][0]
## ugh, should have put this into a class or something..

#print 'china index boundaries',getboundary('China')
ID = np.loadtxt("gpwv3_ntlbndid_ascii_25/glbnds.asc",skiprows=6,dtype='int')
```

```python
def plotcountry(matchstring,pmin=1,pmax=1.e5,logopt=1,fig=None,ax=None,plotcbar=True,plotzoneopt=0,plotzone=[4.,20.,100,1e6]):
  """
  Plot pop density for a particular country
  Hard code exception for continental US, crosses -180 line.
  Use matchstring = '' for hte world.
  """

  myid = getNatID(matchstring)
  xx = np.where(ID != myid)
  if matchstring == '':
    xx = np.where(ID == hdat['NODATA_value'])
    ii, ll = getboundary(matchstring)
  else:
    myid = getNatID(matchstring)
    xx = np.where(ID != myid)
    ii, ll = getboundary(matchstring)          
        
  
  if ax == None and fig == None:
    fig = plt.figure(figsize=[15,15*(ll[2]-ll[3])/(ll[0]-ll[1])])
    ax = fig.add_subplot(111)
  elif ax == None:
    ax = fig.add_subplot(111)    
  
  ptmp = pdens.copy()
  ptmp[xx] = 0. ## set pop outside country boundary to min.
  ptmp = ptmp[ii[0]:ii[1],ii[2]:ii[3]] ## restrict to rectangle around country
  if plotzoneopt == 1:
    xx = np.where((ptmp < plotzone[0]) | (ptmp >= plotzone[-1]))
    ptmp[xx] = 0.
    startval = 1.
    for i in range(len(plotzone)-1):
      xx = np.where((ptmp >= plotzone[i]) & (ptmp < plotzone[i+1]))
      ptmp[xx] = startval
      startval += 1
    #print 'hey beth',ptmp.min(),ptmp.max(), len(plotzone)
  else:
    if ptmp.max() < pmax: pmax = ptmp.max()
    xx = np.where(ptmp > pmax)
    ptmp[xx] = pmax
    xx = np.where(ptmp < pmin)
    ptmp[xx] = pmin
    if logopt == 1:
      ptmp = np.log(ptmp/pmin)/np.log(pmax/pmin)
    #print ptmp.min(), ptmp.max()

  cax = ax.imshow(ptmp,extent=ll)
  if plotcbar == True:
    if logopt == 1:
      cbar = fig.colorbar(cax,ticks=[0.,0.5,1.])
      cbar.ax.set_yticklabels([r'$10^{%.1f}$' % (np.log10(pmin)),r'$10^{%.1f}$' % (0.5*(np.log10(pmin)+np.log10(pmax))),r'$10^{%.1f}$' % (np.log10(pmax))],fontsize=16)  
    elif plotzoneopt == 1:
      cbar = fig.colorbar(cax,ticks=np.arange(1.0,startval+0.01,1.0))  
    else:
      cbar = fig.colorbar(cax)
  return fig, ax
    
def getcountryinfo(matchstring,pzones=[4.,20.,100.]):
  myid = getNatID(matchstring)
  cname = getNatName(matchstring)

  ## want to accomodate the whole world as well.
  if matchstring == '':
    totarea = agrid.sum(); totpop = pcnts.sum()
    cname = 'World'
  else:
    xx = np.where(ID == myid)
    totarea = agrid[xx].sum()
    totpop = pcnts[xx].sum()
  pcntsdist = []
  adist = []
    
  if matchstring == '':
    yy = np.where((pdens < pzones[0]))
  else:
    yy = np.where((ID == myid) & (pdens < pzones[0]))
    
  pcntsdist.append(pcnts[yy].sum())
  adist.append(agrid[yy].sum())
  for i in range(len(pzones)-1):
    if matchstring == '':
      yy = np.where((pdens >= pzones[i]) & (pdens < pzones[i+1]))
    else:
      yy = np.where((ID == myid) & (pdens >= pzones[i]) & (pdens < pzones[i+1]))
    pcntsdist.append(pcnts[yy].sum())
    adist.append(agrid[yy].sum())
  if matchstring == '':
    yy = np.where((pdens > pzones[-1]))
  else:
    yy = np.where((ID == myid) & (pdens > pzones[-1]))
  pcntsdist.append(pcnts[yy].sum())
  adist.append(agrid[yy].sum())
  pzones = np.array(pzones)
  pcntsdist = np.array(pcntsdist)
  adist = np.array(adist)
  outstr = '-'*140+'\n'
  outstr += '%.25s: Population = %.2e, Area = %.2e km^2\n' % (cname,pcntsdist.sum(),adist.sum())
  tmpstr = '< %.2e|' % (pzones[0])
  line2 = 'Pop dens:    |%25s' % (tmpstr)
  line3 = 'Pop (frac):  |'
  line4 = 'Area (frac): |'
  for i in range(len(pzones)-1):
    tmpstr = '%.2e - %.2e|' % (pzones[i], pzones[i+1])
    line2 += '%25s' % (tmpstr)
  tmpstr = '> %.2e |' % (pzones[-1])
  line2 += '%25s' % (tmpstr)
  outstr = outstr + line2 + '\n'  
  

  for i in range(len(pcntsdist)):
    tmpstr1 = '%.2e (%.3f)|' % (pcntsdist[i], pcntsdist[i]/pcntsdist.sum())
    tmpstr2 = '%.2e (%.3f)|' % (adist[i], adist[i]/adist.sum())
    line3 += '%25s' % (tmpstr1)
    line4 += '%25s' % (tmpstr2)
  outstr += line3 + '\n' + line4 + '\n'    
  outstr += '-'*140+'\n\n'      
  
  return pzones, pcntsdist, adist, outstr
```
